```python
kupci_posjek_po_godinama.plot(kind="line",\
                              x="Age",\
                              y="Spending Score (1-100)",\
                              figsize=(10,5),\
                              style="o--b",\
                              legend=True,\
                              title="Srednja vrijednost bododvanja po godinama"
                            )
```

```python
kupci_df.groupby(by="Age", as_index=False).mean(numeric_only=True).plot(\
   kind="line",\
   x="Age",\
   y=["Spending Score (1-100)","Annual Income (k$)"],\
   figsize=(10,5),\
   style=["o--b", "x:g"],\
   legend=True,\
   title="Srednja vrijednost bododvanja po godinama" )
```

```python
#koliko kolona u y toliko i subplotova mora biti
kupci_df.groupby(by="Age", as_index=False).mean(numeric_only=True).plot(\
   kind="line",\
   x="Age",\
   y=["Spending Score (1-100)","Annual Income (k$)"],\
   subplots=True,\
   layout=(1,2),\
   figsize=(10,5),\
   style=["o--b", "x:g"],\
   legend=True,\
   title="Srednja vrijednost bododvanja po godinama" )
```
