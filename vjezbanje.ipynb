Napravite funkciju koja ce racunati median iz liste elemenata

```python
def izracunaj_medijan(lista):
    sortirana_lista = sorted(lista)
    broj_elemenata = len(lista)
    
    if broj_elemenata % 2 != 0:
        return sortirana_lista[broj_elemenata // 2]
    else:
        srednji_desni = sortirana_lista[broj_elemenata // 2]
        srednji_lijevi = sortirana_lista[broj_elemenata // 2 - 1]
        return (srednji_desni + srednji_lijevi) / 2
        
```

```python
izracunaj_medijan([1, 4, 7, 11, 17])
```

```python
izracunaj_medijan([1, 4, 7,6, 11, 17])
```

Napravite funkciju koja ce racunati raspon (range) iz liste elemenata

```python
def izracunaj_raspon(lista):
    if len(lista) == 0:
        return 0
    
    return max(lista) - min(lista)
```

```python
izracunaj_raspon([1, 4,-20, 7, 11, 17])
```

Napravite funkciju koja ce racunati raspon (range) apsolutnih vrijednosti iz liste elemenata

```python
def izracunaj_raspon_od_abs_vrijednosti_liste(lista):
    lista_apsolutnih_vrijednosti = []
    
    for elem in lista:
        lista_apsolutnih_vrijednosti.append(abs(elem))
        
    return izracunaj_raspon(lista_apsolutnih_vrijednosti)
```

```python
izracunaj_raspon_od_abs_vrijednosti_liste([1, 4,-20, 7, 11, 17])
```

```python
def izracunaj_raspon_od_abs_vrijednosti_liste_2(lista):
    lista_apsolutnih_vrijednosti = [abs(elem) for elem in lista]
        
    return izracunaj_raspon(lista_apsolutnih_vrijednosti)
```

```python
izracunaj_raspon_od_abs_vrijednosti_liste_2([1, 4,-20, 7, 11, 17])
```

```python
def izracunaj_raspon_od_abs_vrijednosti_liste_3(lista):
    lista_apsolutnih_vrijednosti = list(map(abs, lista))
        
    return izracunaj_raspon(lista_apsolutnih_vrijednosti)
```

Napisati funkciju koja racuna frekvencije za svaki element (pandasov value counts)

```python
def izracunaj_fekvencije(lista):
    
    if not lista:
        return dict()
    
    razliciti_elementi = set(lista)
    brojaci = {}
    
    for elem in razliciti_elementi:
        brojaci[elem] = 0
        
    for element in lista:
        brojaci[element] += 1
        
    return brojaci
```

```python
brojevi = [5, 2, 9, 1, 7, 2, 2, 5, 5, 5]
izracunaj_fekvencije(brojevi)
```

```python
def izracunaj_jedinstvene_vrijednosti(lista):
    return list(set(lista))
```

```python
brojevi = [5, 2, 9, 1, 7, 2, 2, 5, 5, 5]
izracunaj_jedinstvene_vrijednosti(brojevi)
```

Izracunaj mod od liste elemenata

```python
def izracunaj_mod(lista):
    if not lista:
        return None
    
    frekvencije = izracunaj_fekvencije(lista)
    
    mod_liste = None
    maksimalan_vrijednost = 0
    
    for jedinstvena_vrijednost, broj_pojavljivanja in frekvencije.items():
        
        if broj_pojavljivanja > maksimalan_vrijednost:
            mod_liste = jedinstvena_vrijednost
            maksimalan_vrijednost = broj_pojavljivanja
            
    return mod_liste
    
```

```python
brojevi = [5, 2, 9, 1, 7, 2, 2, 5, 5, 5, 7, 7, 7, 7, 7,7 ]
izracunaj_mod(brojevi)
```

```python
def izracunaj_vise_modova(lista):
    if not lista:
        return None
    
    frekvencije = izracunaj_fekvencije(lista)
    
    mod_liste = []
    maksimalan_vrijednost = 0
    
    for jedinstvena_vrijednost, broj_pojavljivanja in frekvencije.items():
        if broj_pojavljivanja > maksimalan_vrijednost:
            mod_liste = [jedinstvena_vrijednost]
            maksimalan_vrijednost = broj_pojavljivanja
            
        elif broj_pojavljivanja == maksimalan_vrijednost:
            mod_liste.append(jedinstvena_vrijednost)
            
        
    return mod_liste
```

```python
brojevi = [5, 2, 9, 1, 7, 2, 2, 5, 5, 5, 7, 7,7, 9, 9, 9]
izracunaj_vise_modova(brojevi)
```

Napravi funkciju koja racuna kvantil (1, 2, 3, 4) 1 -> Q1, 2->Q2 od liste elemenata

```python
def izracunaj_kvantil(lista, kvantil):
    #Q1 25%, Q2 50% --> medijan, Q3 75%, Q4 100%
    
    if not lista:
        return None
    
    kvantili = {1: 0.25, 2: 0.5, 3:0.75, 4:1.0}
    
    if kvantil not in kvantili:
        return None
    
    broj_elemenata = len(lista)
    sortirana_lista = sorted(lista)
    
    index = int((broj_elemenata - 1) * kvantili[kvantil])
    
    return sortirana_lista[index]
    
```

```python
brojevi = [1, 5,  6, 7,   11, 19,   24, 300]
print(izracunaj_kvantil(brojevi, 1))
print(izracunaj_kvantil(brojevi, 2))
print(izracunaj_kvantil(brojevi, 3))
print(izracunaj_kvantil(brojevi, 4))

```

```python
brojevi = [1, 5,     11, 19,   24, 300, 6, 7,]
print(izracunaj_kvantil(brojevi, 1))
print(izracunaj_kvantil(brojevi, 2))
print(izracunaj_kvantil(brojevi, 3))
print(izracunaj_kvantil(brojevi, 4))
```
