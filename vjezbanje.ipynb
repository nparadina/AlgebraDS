```python
def median(list):
    list.sort()
    n=len(list)
    if n==0:
        median=-1
    elif n%2==0:
        median=(list[n//2]+list[n//2-1])/2
    else:
        median=list[n//2+1]

    return median
```

```python
def izracunaj_range(list):
    for i in range(0,len(list)):
        if list[i]<0:
            list[i]=abs(list[i])
    return max(list)-min(list)
```

```python
list=[-1,2,5,6]
izracunaj_range(list)
```

```python

def izracunaj_frekvencije (list):
    dict={}
    for elem in list:
        if elem in dict.keys():
            dict[elem]=dict[elem]+1
        else:
            dict[elem]=1

    return dict

```

```python
list=[-1,2,5,6,6,6,-1, 3]
izracunaj_frekvencije(list)
```

```python
def izracunaj_mod(list):
    brojac=izracunaj_frekvencije(list)
    mod_vrijednost=max(brojac.values())
    mod_kljuc=[]
    for key,value in brojac.items():
        if value==mod_vrijednost:
            mod_kljuc.append(key)
    return mod_kljuc
```

```python
list=[-1,2,5,6,6,6,-1, 3,3,3]
izracunaj_mod(list)
```
