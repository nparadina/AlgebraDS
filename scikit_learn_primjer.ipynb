```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# 2. ocitavanej podataka
iris = load_iris()
X = iris.data
y = iris.target

print(X)
print(y)
```

```python
iris.target_names
```

```python
#3. podjela na traning skup i test skup 
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#4. kreiraj model
model = ComplementNB()

#5. treniraj model
model.fit(X_train, y_train)

#6. predikcija modela
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=iris.target_names)

print(f"Tocnost modela {accuracy:.2f}")
print(f"\nIzvjestaj klasifikacija: \n{report}")
```

```python
from sklearn.cluster import KMeans
```

```python
X = np.array([[5.1, 3.5], [4.9, 3.0], [4.7, 3.2], [6.6, 4.6], [6.3, 4.4], [6.8, 4.8], [5.0, 3.6], [6.4, 4.5], [5.3, 3.9], [6.5, 4.7]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
model = KMeans(n_clusters=2, random_state=100)

model.fit(X)

y_pred = model.predict(X)
cetroidi = model.cluster_centers_

plt.figure(figsize=(8,8))

plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap="viridis", label="Data Points")
plt.scatter(cetroidi[:, 0], cetroidi[:, 1], c="red", s=200, marker="X", label="Centroidi")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.title("K-Means algoritam, grupiranja")

plt.show()
```

```python
model = KMeans(n_clusters=4, random_state=42)

model.fit(X)

y_pred = model.predict(X)
cetroidi = model.cluster_centers_

plt.figure(figsize=(8,8))

plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap="viridis", label="Data Points")
plt.scatter(cetroidi[:, 0], cetroidi[:, 1], c="red", s=200, marker="X", label="Centroidi")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.title("K-Means algoritam, grupiranja")

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

X = 6 * np.random.rand(100,1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(100,1)

plt.scatter(X, y)
plt.title("Umjetni podatci")
plt.xlabel("Kolona 1")
plt.ylabel("Kolona 2")
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)

sort_idx = np.argsort(X, axis=0).flatten()
X_sorted = X[sort_idx]
y_pred_sorted = y_pred[sort_idx]

plt.scatter(X, y, color="blue", label="Pravi podatci")
# plt.plot(X_sorted, y_pred_sorted, color="red", linewidth=2, label="Linearna regresija")
plt.plot(X, y_pred, color="red", linewidth=2, label="Linearna regresija")

plt.title("Umjetni podatci i linearna regresija")
plt.xlabel("Kolona 1")
plt.ylabel("Kolona 2")
plt.legend()
plt.show()
```

```python
mean_squared_error(y, y_pred)
```

```python
r2_score(y, y_pred)
```
