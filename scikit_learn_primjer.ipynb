```python
#Loadanje podataka, ovo moze doci i ne iz scikit learn-a, nego iz realnog zivota
from sklearn.datasets import load_iris
# load metode za kreiranje testnog i training seta-a podataka iz nasih podataka
from sklearn.model_selection import train_test_split
# loadanje modela koji nam treba, tu moze ici RandomForest, NeuralNetworks itd...
from sklearn.naive_bayes import ComplementNB
#loadanje metrika za ocjenu rada modela
from sklearn.metrics import accuracy_score, classification_report
```

```python
#krieanje ulazne matrica X i izlaznih rezultata y. Da se ne uzima iz iris-a, samo bismo morali kreirati X i y. X je veliko jer je matrica
iris=load_iris()
X=iris.data
y=iris.target

print(X)
print(y)
```

```python
iris.target_names
```

```python
# Kreiranje iz svih podataka testnog i trening skupa
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3, random_state=42) # random_state je namjerno stavljen na 42 jer s time se moze inicijalno stanje deterministicki postavit
```

```python
#inicijalizacija modela
model=ComplementNB()
print(type(model))
#treninranje modela
model.fit(X_train, y_train)
```

```python
#predikcija
y_pred=model.predict(X_test)

```

```python
#usporedba rezultata po metrikama
accuracy=accuracy_score(y_test,y_pred)
print(f' Tocnost modela {accuracy: .2f}')
report=classification_report(y_test,y_pred, target_names=iris.target_names)
print(f'\n Izvjestaj: \n {report}')
```

```python
from  sklearn.cluster import KMeans
 
```

```python
X=np.array([[5.1, 3.5],[4.9,3.0],[4.7, 3.2],[6.6,4.6],[6.3,4.4],[6.8,4.8],[5.0,3.6],[6.4,4.5],[5.3,3.9],[6.5,4.75]])
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
model_kmeans=KMeans(n_clusters=2, random_state=42)

model_kmeans.fit(X)
```

```python
y_pred=model_kmeans.predict(X)
centroidi= model_kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[:,0], X[:,1], c=y_pred, s=50, cmap="viridis", label="Data Points")
plt.scatter(centroidi[:,0], centroidi[:,1], c="red", s=200, marker="X", label="Centroidi")

plt.show()
```
