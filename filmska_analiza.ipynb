```python
def directors_statistics(director_names, number_of_films, revenues, votes, popualrities):
    categories = ["Num od film", "Revenue", "Num of votes", "Popularity"]
    colors = ["r", "g", "b", "m", "y", "c"]
    num_categories = len(categories)
    
    angles = np.linspace(0, 2*np.pi, num_categories, endpoint=False)
    
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    plt.title("Redatelji", fontsize=18)
    
    for director_name, number_of_film, revenue, num_votes, popularity in zip(director_names, number_of_films, revenues, votes, popualrities):
        values = [number_of_film, revenue, num_votes, popularity]
        ax.plot(np.concatenate([angles, [angles[0]]]), np.concatenate([values, [values[0]]]), "o-", label=director_name)
        ax.fill(np.concatenate([angles, [angles[0]]]), np.concatenate([values, [values[0]]]), alpha=0.25)
        
    ax.set_thetagrids(angles * 180/np.pi, categories)
    ax.set_rlabel_position(0)
    

    
    plt.legend(loc="upper right", bbox_to_anchor=(0.1, 0.1))
    plt.show()
        
```
